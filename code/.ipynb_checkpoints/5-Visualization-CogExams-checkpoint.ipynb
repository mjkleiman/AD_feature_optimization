```python
import numpy as np
import pandas as pd
from glob import glob
import seaborn as sns
import matplotlib.pyplot as plt

filenames = glob('../models/outputs/CExam_*.csv')
dataframes = [pd.read_csv(f) for f in filenames]

df = pd.concat(dataframes)
```

```python
df.groupby('set')['Sensitivity','Specificity','PPV','NPV','Accuracy','F1_Score'].mean()
```

```python
sns.barplot(x='Sensitivity', y='set', data=df).set(xlim=(.4,None))
```

```python
sns.barplot(x='Specificity', y='set', data=df).set(xlim=(.5,None))
```

```python
sns.barplot(x='PPV', y='set', data=df).set(xlim=(.6,None))
```

```python
sns.barplot(x='NPV', y='set', data=df).set(xlim=(.6,None))
```

```python
sns.barplot(x='Accuracy', y='set', data=df).set(xlim=(.6,None))
```

```python
sns.barplot(x='F1_Score', y='set', data=df).set(xlim=(.6,None))
```
