```python
from warnings import simplefilter
simplefilter(action='ignore', category=FutureWarning)
import numpy as np
import pandas as pd
import pingouin as pg
from glob import glob
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

filenames = glob('../models/outputs/*.csv')
dataframes = [pd.read_csv(f) for f in filenames]

df = pd.concat(dataframes)

df.loc[df['set'].str.contains('2class'), '2v3'] = 2
df.loc[~df['set'].str.contains('2class'), '2v3'] = 3
df.loc[df['set'].str.contains('Imp'), '2v3'] = 2


df.loc[df['set'].str.contains('CExam'), 'CExam'] = 1
df.loc[~df['set'].str.contains('CExam'), 'CExam'] = 0

df_cexams = df[df['CExam'] == 1]
# df.to_csv('../models/allmodels_output.csv', index=False)
```

```python
df_set = df.groupby('set')['Sensitivity','Specificity','PPV','NPV','Accuracy','F1_Score'].mean().reset_index()

df_set.loc[df_set['set'].str.contains('2class'), '2v3'] = 2
df_set.loc[~df_set['set'].str.contains('2class'), '2v3'] = 3
df_set.loc[df_set['set'].str.contains('Imp'), '2v3'] = 2

df_set
```

```python
display(df.groupby('CExam')['Sensitivity','Specificity','PPV','NPV','Accuracy','F1_Score'].mean())
display(df.groupby('2v3')['Sensitivity','Specificity','PPV','NPV','Accuracy','F1_Score'].mean())
```

```python
display(df_cexams.groupby('set')['Sensitivity','Specificity','PPV','NPV','Accuracy','F1_Score'].mean())
display(df_cexams.groupby('2v3')['Sensitivity','Specificity','PPV','NPV','Accuracy','F1_Score'].mean())
```

```python
df.loc[(df['CExam'] == 1) & (df['2v3'] == 3.0)].describe()
```

# Sensitivity 

```python
display(df.groupby('CExam')['Sensitivity'].describe())
aov_sens = pg.anova(data=df, dv='Sensitivity', between='CExam')
display(aov_sens)
phoc_sens = pg.pairwise_tukey(data=df, dv='Sensitivity', between='CExam')
display(phoc_sens)
```

```python
display(df.groupby('set')['Sensitivity'].describe())
aov_sens = pg.anova(data=df, dv='Sensitivity', between='set')
display(aov_sens)  
phoc_sens = pg.pairwise_tukey(data=df, dv='Sensitivity', between='set')
display(phoc_sens)
```

```python
df['TwoVThree'] = df['2v3']
aov_sens = df.anova(dv='Sensitivity', between=['CExam','TwoVThree'])
display(aov_sens)
```

## CExams Only

```python
aov_levene = pg.homoscedasticity(data=df_cexams, dv='Sensitivity', group='set', method='levene', alpha=0.05)
display(aov_levene)
```

```python
display(df_cexams.groupby('set')['Sensitivity'].describe())
aov_sens = pg.anova(data=df_cexams, dv='Sensitivity', between='set')
display(aov_sens)
phoc_sens = pg.pairwise_tukey(data=df_cexams, dv='Sensitivity', between='set')
display(phoc_sens)
```
