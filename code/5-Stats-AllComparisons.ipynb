```python
from warnings import simplefilter
simplefilter(action='ignore', category=FutureWarning)
import numpy as np
import pandas as pd
import pingouin as pg
from glob import glob
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

filenames = glob('../models/outputs/CDR_*.csv')
dataframes = [pd.read_csv(f) for f in filenames]

df = pd.concat(dataframes)

df['set'].replace(
    {'CDR_MultiC_AllFeatures_400':'3-class AllFeatures',
    'CDR_UMCN_BorutaECOGFAQ_400':'3-class Manual-4+ECOG/FAQ MCN',
    'CDR_MultiC_BorutaSHAP_400':'3-class Boruta-SHAP',
    'CDR_Imp_Boruta8_400':'2-class Manual-8',
    'CDR_UMCN_BorutaSHAP_400':'3-class Boruta-SHAP MCN',
    'CDR_Imp_UMCN_B_EFAQ_400':'2-class Manual-8+ECOG/FAQ MCN',
    'CDR_UMCN_AllFeatures_400':'3-class All-Features MCN',
    'CDR_Imp_UMCN_Boruta4-8_400':'2-class Manual-4/8 MCN',
    'CDR_MultiC_BorutaECOGFAQ_400':'3-class Manual-4+ECOG/FAQ',
    'CDR_Imp_AllFeatures_400':'2-class All-Features',
    'CDR_Imp_Boruta4_400':'2-class Manual-4',
    'CDR_UMCN_Boruta4_400':'3-class Manual-4/8+ECOG MCN',
    'CDR_Imp_BorutaSHAP_400':'2-class Boruta-SHAP',
    'CDR_Imp_B_EFAQ_400':'2-class Manual-4+ECOG/FAQ'}, inplace=True)
```

```python
df_desc = df.groupby('set')['Sensitivity','PPV','Specificity','NPV','Accuracy','F1_Score'].mean()
# df_desc.to_excel('../reports/figures/table2.xlsx')
df_desc
```

```python
df.groupby('set')['Sensitivity','Specificity','PPV','NPV'].var().describe()
```

# Specificity

### Levene's test:

```python
aov_levene = pg.homoscedasticity(data=df, dv='Specificity', group='set', method='levene', alpha=0.05)
display(aov_levene)
```

```python
display(df.groupby('set')['Specificity'].describe())
aov_sens = pg.anova(data=df, dv='Specificity', between='set')
display(aov_sens)
phoc_sens = pg.pairwise_tukey(data=df, dv='Specificity', between='set')
display(phoc_sens)
```
